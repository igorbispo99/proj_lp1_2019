```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
def read_labels(txt_name):
    f = open(txt_name, "r")
    
    y_pred = []    
    y_real = []
    
    for line in f:
        y_p, y_r = line.split()
        y_pred.append(y_p)
        y_real.append(y_r)
    
    return np.array(y_pred), np.array(y_real)
```

```python


import matplotlib.pyplot as plt
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

```

# Mnist

```python
y_p_mnist, y_r_mnist = read_labels("out_mnist.txt")
```

```python
print(classification_report(y_r_mnist, y_p_mnist))
print(accuracy_score(y_r_mnist, y_p_mnist))

plot_confusion_matrix(confusion_matrix(y_r_mnist, y_p_mnist), np.arange(0, 10))
```
